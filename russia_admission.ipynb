### The total applications per year in the data set. 

```python
%%sql 

SELECT Year, SUM(`Number of Applications`) as Total_No_Application
FROM university_admission_russia
GROUP BY Year
```

```python
%%sql 

SELECT Year, SUM(`Number of Applications`) as Total_No_Application
FROM university_admission_russia
GROUP BY Year
ORDER BY Total_No_Application DESC
```

The above is the total number of application per year in the Uni admission. The table is arranged with the highest to the lowest (DESC ORDER). The highest number of applications were received in 2023 and the lowest received in 2014.

```python
%%sql 

SELECT `Branches of Science` as Faculty, SUM(`Number of Applications`) as Total_No_Application
FROM university_admission_russia
GROUP BY Faculty
ORDER BY Total_No_Application DESC
```

The ten years (2014-2023) total number of applications in the Uni per Faculty (Branches of Science) arranged in a Desc ORDER. The analysis revealed that Engineering and Technology had the highest number of applications while Military Science had the least.

```python
%%sql 

SELECT Year, `Branches of Science` as Faculty, SUM(`Number of Students`) as Total_Admission
FROM university_admission_russia
GROUP BY Year, Faculty
ORDER BY Total_Admission DESC
```

```python
%%sql 

SELECT Year, SUM(`Number of Applications`) as Total_No_Application, SUM(`Number of Students`) as Total_Admission
FROM university_admission_russia
GROUP BY Year
ORDER BY Total_No_Application, Total_Admission 
```

```python
%%sql 

SELECT Year, SUM(`Number of Applications`) as Total_No_Application, SUM(`Number of Students`) as Total_Admission, (SUM(`Number of Applications`) -  SUM(`Number of Students`)) as UnSuccessful_Applicatns
FROM university_admission_russia
GROUP BY Year
ORDER BY Total_No_Application, Total_Admission 
```

```python
%%sql 

SELECT 
    Year, 
    SUM(`Number of Applications`) as Total_No_Application, 
    SUM(`Number of Students`) as Total_Admission, 
    (SUM(`Number of Applications`) -  SUM(`Number of Students`)) as UnSuccessful_Applicatns,
    ((SUM(`Number of Students`)/SUM(`Number of Applications`))*100) as percentage_Acceptance_Rate,
    ((SUM(`Number of Applications`) -  SUM(`Number of Students`))/SUM(`Number of Applications`))*100) as Percentage_Rej_Rate,
FROM university_admission_russia
GROUP BY Year
ORDER BY Total_No_Application, Total_Admission 
```

```python
%%sql 

SELECT 
    Year, 
    SUM(`Number of Applications`) as Total_No_Application, 
    SUM(`Number of Students`) as Total_Admission, 
    (SUM(`Number of Applications`) -  SUM(`Number of Students`)) as UnSuccessful_Applicatns,
    ((SUM(`Number of Students`)/SUM(`Number of Applications`))*100) as Percent_Accept_Rate,
    (((SUM(`Number of Applications`) -  SUM(`Number of Students`))/SUM(`Number of Applications`))*100) as Percent_Rej_Rate
FROM university_admission_russia
GROUP BY Year
ORDER BY Total_No_Application, Total_Admission
```

The above is a tabular distri. of the Yr, No. of Appl., Admissions, Percentage Acceptance Rate, and the Percentage Rejection Rate in the Uni for 10 years (2014 to 2023)
